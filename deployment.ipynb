```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("harsh098mumbai/lyrics_generator_asg_gpt2")

model = AutoModelForCausalLM.from_pretrained("harsh098mumbai/lyrics_generator_asg_gpt2",pad_token_id=tokenizer.eos_token_id)
```

```python
def generate_text(prompt):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output = model.generate(input_ids, max_length=30, num_return_sequences=1)
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text

def generate_text_interface(prompt):
    generated_text = generate_text(prompt)
    return generated_text
```

```python
import gradio as gr
```

```python
interface = gr.Interface(
    fn=generate_text_interface,
    inputs=gr.inputs.Textbox(label="Enter Prompt"),
    outputs=gr.outputs.Textbox(label="Generated Text Output"),
    title="Text Generation",
    description="Type some lyrics.",
)
interface.launch()
```
